```python
from collections import Counter
import csv
def load_monitor_label(monitor_label_path):
    # monitor_label_path = './dataset/monitor_label.csv'
    label_cnter = Counter()
    with open(monitor_label_path, newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')

        sub_ls, obj_ls, label_ls = [], [], []
        for ind, row in enumerate(spamreader):
#             print(row)
            if ind == 0:
                continue
            sub, obj, label = row[-1].split(',')
#             print(row[-1].split(','))
#             print(sub, obj, label)
            label_cnter.update([sub.split('//')[0]+','+obj.split('//')[0]])
            sub_ls.append(sub)
            obj_ls.append(obj)
            label_ls.append(int(label))

        df = pd.DataFrame.from_dict({'subject1':sub_ls, 'subject2':obj_ls, 'label':label_ls})

    return df
```

```python
def entry_parse(input):
#     print(type(input))
#     exit()
    if type(input) is str:
        result = input.replace('\n', ' ').replace('\r', ' ')
#         print(result)
        return result
    else:
        return input

    
```

```python
def parse_source(input_str):
    
    return input_str.split('//')[0]


def sample_extract(df, k):
    df_p = df[df['label'] == 1]
    df_n = df[df['label'] == 0]
    
    df_p_sample = df_p.sample(k)
    df_n_sample = df_n.sample(k)
    
    df_kshot = pd.concat([df_p_sample, df_n_sample])
    df_test = df.drop(df_kshot.index)
    
    return df_kshot, df_test
```
